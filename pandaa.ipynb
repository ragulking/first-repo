```python
#remove the quter lier
#Remove NaN values.

#Visualize the data before and after removal.

#Z-score method (for normally distributed data).

#IQR (Interquartile range) method for more robust data.
```

```python
Q1 = df["City_Delhi"].quantile(0.25)
Q3 = df["City_Delhi"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[df["City_Delhi"].between(lower_bound, upper_bound)]
```
