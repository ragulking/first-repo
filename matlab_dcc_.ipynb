```python
x=[0,1,2,3,4,5]
y=[1,2,3,4,5,6]
a = np.arange(0,4.5,0.5)
plt.title('the increasing')
plt.xlabel('x values')
plt.ylabel('2X values')
plt.plot(x,y,label ='2x' ,  color = 'y',linewidth = 2,marker ='.' , linestyle  = '--',markersize = 10 , markeredgecolor = 'red')
plt.legend()
plt.plot(a,a*2)
```
