```python
# Find the most repeated (mode) value
 # Replace null values with it
values = ['Customer', 'Product', 'Region']

for i in values:
    most_common_name = uncleaned_data[i].mode()[0]
    uncleaned_data[i].fillna(most_common_name, inplace=True)



```
